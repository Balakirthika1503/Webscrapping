```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
import pyautogui
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
```

```python
PATH="C:\Program Files (x86)\chromedriver.exe"
driver=webdriver.Chrome(PATH)
driver.get("https://www.gfxtra31.com/programming-books/")
driver.maximize_window()
```

```python
Prod_Name,Prod_URL,Filehoster_URL = [],[],[] 
for i1 in range(24):
    #Prod_Name.append
    Prod_Name.append(driver.execute_script('return document.getElementsByClassName("baslik")['+str(i1)+'].innerText').split('\n')[0])
    href = driver.execute_script('return document.getElementsByClassName("baslik")['+str(i1)+'].innerHTML').split('href=')[1].split('>')[0].replace('"','')
    #Prod_URL.append
    Prod_URL.append(href)
    driver.execute_script('window.open();')
    driver.switch_to.window(driver.window_handles[1])
    driver.get(href)
    time.sleep(2)
    """ COLLECT SOUP """
    content = driver.page_source
    soup = BeautifulSoup(content, "html.parser")

    demo = str(soup.findAll('div',attrs={'style':'display:inline;'})[0]).split('display:inline;')[0].split('div id')[1].split('style')[0].replace(' ','')
    try:
        try:            
            driver.find_element_by_xpath('//*[@id'+str(demo)+']/p[2]/a').click()
        except Exception as e:
            try:
                driver.find_element_by_xpath('//*[@id'+str(demo)+']/p[3]/a').click()
            except Exception as e:
                try:
                    driver.find_element_by_xpath('//*[@id'+str(demo)+']/p[7]/a').click()
                except Exception as e:
                    try:
                        driver.find_element_by_xpath('//*[@id'+str(demo)+']/p[16]/a').click()
                    except Exception as e:
                        try:
                            driver.find_element_by_xpath('//*[@id'+str(demo)+']/p[37]/a/img').click()
                        except Exception as e:
                            try:
                                driver.find_element_by_xpath('//*[@id'+str(demo)+']/p[37]/a/img').click()
                            except Exception as e:
                                try:
                                    driver.find_element_by_xpath('//*[@id'+str(demo)+']/p').click()
                                except Exception as e:
                                    print('NA')
                                
    except Exception as e:
        try:
            driver.find_element_by_xpath('//*[@id'+str(demo)+']/p[8]/a').click()
        except Exception as e:
            try:
                driver.find_element_by_xpath('//*[@id'+str(demo)+']/a').click()
            except Exception as e:
                try:
                    driver.find_element_by_xpath('//*[@id'+str(demo)+']/ul/li[5]/p[2]/a').click()
                except Exception as e:
                    try:
                        driver.find_element_by_xpath('//*[@id'+str(demo)+']/ul/li[4]/p[2]/a').click()
                    except Exception as e:
                        print('NA')
                    
                        
                        
    
    
    Filehoster_URL.append(driver.current_url)
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
        
        
        
    
```

```python
data = pd.DataFrame({'Product_Name':Prod_Name,'Product_URL':Prod_URL,'File_Hoster_URL':Filehoster_URL})

data
```

```python
data.to_csv('books.csv')
```
